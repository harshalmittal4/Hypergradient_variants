```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
sgd_adam = pd.read_csv('test/mlp/+1e-03_+1e-03/sgd_HdAdam.csv')
sgd_hd = pd.read_csv('test/mlp/+1e-03_+1e-03/sgd_Hd.csv')
sgd_sgdn = pd.read_csv('test/mlp/+1e-03_+1e-03/sgd_Hdmomentum.csv')
sgd = pd.read_csv('test/mlp/+1e-03_+1e-06/sgd.csv')

sgdn_hd = pd.read_csv('test/mlp/+1e-03_+1e-03/sgdn_Hd.csv')
sgdn_sgdn = pd.read_csv('test/mlp/+1e-03_+1e-03/sgdn_Hdmomentum.csv')
sgdn_adam = pd.read_csv('test/mlp/+1e-03_+1e-03/sgdn_HdAdam.csv')
sgdn = pd.read_csv('test/mlp/+1e-03_+1e-06/sgdn.csv')

adam = pd.read_csv('test/mlp/+1e-03_+1e-06/adam.csv')
adam_hd = pd.read_csv('test/mlp/+1e-03_+1e-07/adam_Hd.csv')
adam_adam = pd.read_csv('test/mlp/+1e-03_+1e-07/adam_HdAdam.csv')
adam_sgdn = pd.read_csv('test/mlp/+1e-03_+1e-07/adam_Hdmomentum.csv')
```

```python
sgd_adam.head()
```

```python
data = pd.DataFrame()
data['sgd'] = sgd['LossEpoch']
data['sgd_HdAdam'] = sgd_adam['LossEpoch']
data['sgd_Hd'] = sgd_hd['LossEpoch']
data['sgd_Hdmomentum'] = sgd_sgdn['LossEpoch']
data['sgdn_HdAdam'] = sgdn_adam['LossEpoch']
data['sgdn_Hd'] = sgdn_hd['LossEpoch']
data['sgdn_Hdmomentum'] = sgdn_sgdn['LossEpoch']
data['sgdn'] = sgdn['LossEpoch']
data['adam_HdAdam'] = adam_adam['LossEpoch']
data['adam_Hdmomentum'] = adam_sgdn['LossEpoch']
data['adam'] = adam['LossEpoch']
data['adam_Hd'] = adam_hd['LossEpoch']
```

```python
data.head()
```

```python
data_ = data.dropna(axis=0, how='all')
```

```python
data_.head()
```

```python
len(data_)
```

```python
a = list(data.columns)
```

```python
a
```

```python
a
config = {}
config['sgd'] = {'c':'gold', 'linestyle':':'}
config['sgd_HdAdam'] = {'c':'darkred', 'linestyle':':'}
config['sgd_Hd'] = {'c':'black', 'linestyle':':'}
config['sgd_Hdmomentum'] = {'c':'blue', 'linestyle':':'}
config['sgdn_HdAdam'] = {'c':'green', 'linestyle':'--'}
config['sgdn_Hd'] = {'c':'teal', 'linestyle':'--'}
config['sgdn_Hdmomentum'] = {'c':'purple', 'linestyle':'--'}
config['sgdn'] = {'c':'gold', 'linestyle':'--'}
config['adam_HdAdam'] = {'c':'darkturquoise', 'linestyle':'-'}
config['adam_Hdmomentum'] = {'c':'lime','linestyle':'-'}
config['adam'] = {'c':'red', 'linestyle':'-'}
config['adam_Hd'] = {'c':'chocolate', 'linestyle':'-'} 
data_ = data_.reset_index()
```

```python
plt.figure()
for optimizer in a:
    if optimizer in [ 'sgdn_HdAdam',
 'sgdn_Hd',
 'sgdn_Hdmomentum',
 'sgdn']:
        data_[optimizer].plot(color = config[optimizer]['c'], linestyle = config[optimizer]['linestyle'])
plt.legend(loc = 'best')
plt.xlabel('Epochs')
plt.ylabel('Validation Loss')
plt.ylim(top=0.4)

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)

#plt.show()
plt.savefig('mlp_Train_Loss_sgdnVar.pdf', bbox_inches='tight')
```
