```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
sgd_adam = pd.read_csv('test/vgg/+1e-03_+1e-03/opSgd_lopAdam.csv')
sgd_hd = pd.read_csv('test/vgg/+1e-03_+1e-03/opSgd_lopHd.csv')
sgd_sgdn = pd.read_csv('test/vgg/+1e-03_+1e-03/opSgd_lopSgdn.csv')
sgd = pd.read_csv('test/vgg/+1e-03_+1e-06/sgd.csv')

sgdn_hd = pd.read_csv('test/vgg/+1e-03_+1e-03/opSgdn_lopHd.csv')
sgdn_sgdn = pd.read_csv('test/vgg/+1e-03_+1e-03/opSgdn_lopSgdn.csv')
#sgdn_adam = pd.read_csv('test/vgg/+1e-03_+1e-03/opSgdn_lopAdam.csv') missing
sgdn = pd.read_csv('test/vgg/+1e-03_+1e-06/sgdn.csv')

adam = pd.read_csv('test/vgg/+1e-03_+1e-06/adam.csv')
adam_hd = pd.read_csv('test/vgg/+1e-03_+1e-08/opAdam_lopHd.csv')
adam_adam = pd.read_csv('test/vgg/+1e-03_+1e-08/opAdam_lopAdam.csv')
adam_sgdn = pd.read_csv('test/vgg/+1e-03_+1e-08/opAdam_lopSgdn.csv')
```

```python
data = pd.DataFrame()
data['opSgd_lopAdam'] = sgd_adam['ValidLossEpoch']
data['opSgd_lopHd'] = sgd_hd['ValidLossEpoch']
data['opSgd_lopSgdn'] = sgd_sgdn['ValidLossEpoch']
data['sgd'] = sgd['ValidLossEpoch']
#data['opSgdn_lopAdam'] = sgdn_adam['ValidLossEpoch']
data['opSgdn_lopHd'] = sgdn_hd['ValidLossEpoch']
data['opSgdn_lopSgdn'] = sgdn_sgdn['ValidLossEpoch']
data['sgdn'] = sgdn['ValidLossEpoch']
data['opAdam_lopAdam'] = adam_adam['ValidLossEpoch']
data['opAdam_lopSgdn'] = adam_sgdn['ValidLossEpoch']
data['adam'] = adam['ValidLossEpoch']
data['opAdam_lopHd'] = adam_hd['ValidLossEpoch']
```

```python
data_ = data.dropna(axis=0, how='all')
```

```python
data_.reset_index(inplace = True)
```

```python
config = {}
config['opSgd_lopAdam'] = {'c':'darkred', 'linestyle':':'}
config['opSgd_lopHd'] = {'c':'black', 'linestyle':':'}
config['opSgd_lopSgdn'] = {'c':'blue', 'linestyle':':'}
config['sgd'] = {'c':'lime', 'linestyle':':'}
#config['opSgdn_lopAdam'] = {'c':'green', 'linestyle':'--'}
config['opSgdn_lopHd'] = {'c':'teal', 'linestyle':'--'}
config['opSgdn_lopSgdn'] = {'c':'purple', 'linestyle':'--'}
config['sgdn'] = {'c':'gold', 'linestyle':'--'}
config['opAdam_lopAdam'] = {'c':'darkturquoise', 'linestyle':'-'}
config['opAdam_lopSgdn'] = {'c':'lime','linestyle':'-'}
config['adam'] = {'c':'red', 'linestyle':'-'}
config['opAdam_lopHd'] = {'c':'chocolate', 'linestyle':'-'} 
```

```python
a = list(data.columns)
plt.figure()
for optimizer in a:
    data_[optimizer].plot(color = config[optimizer]['c'], linestyle = config[optimizer]['linestyle'])
plt.legend(loc = 'best')
plt.xlabel('Epochs')
plt.ylabel('Validation Loss')
plt.ylim(bottom= 0.25)

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)

#plt.show()
plt.savefig('Vgg_Valid_Loss.pdf', bbox_inches='tight')
```
