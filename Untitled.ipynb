```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test1/logreg/+1e-03_+1e-03/adam_hd_nag_epochs3.csv')
```

```python
df.head()
```

```python
df['Alpha'][1].find('tensor')
```

```python
iteration = []
loss = []
alpha = []
for i in range(len(df)):
    try:
        iteration.append(int(df['Iteration'][i]))
        loss.append(float(df['Loss'][i][7:-1]))
        if df['Alpha'][i].find('tensor') < 0:
            alpha.append(float(df['Alpha'][i]))
        else:
            alpha.append(float(df['Alpha'][i][7:-1]))
            
        
    except:
        continue
```

```python
len(iteration)
```

```python
plt.figure(figsize=(50,25))
plt.plot(iteration, loss)
plt.savefig('HDplots/loss.png')
```

```python
plt.figure(figsize=(50,25))
plt.plot(iteration, alpha)
plt.show("HDplots/alpha.png")
```
