```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_sgd_hd_adam = pd.read_csv('test2/logreg/+1e-03_+1e-03/sgd_hd_adam.csv')
df_sgd_hd_nag = pd.read_csv('test2/logreg/+1e-03_+1e-03/sgd_hd_nag.csv')
df_sgd_hd = pd.read_csv('test2/logreg/+1e-03_+1e-03/sgd_hd.csv')
df_sgd = pd.read_csv('test2/logreg/+1e-03_+1e-03/sgd.csv')
```

```python
iteration = []
loss_sgd_hd_adam = []
loss_sgd_hd_nag = []
loss_sgd_hd = []
loss_sgd = []
alpha_sgd_hd_adam = []
alpha_sgd_hd_nag = []
alpha_sgd_hd = []
alpha_sgd = []

for i in range(max(len(df_sgd_hd_adam),len(df_sgd_hd_nag),len(df_sgd_hd), len(df_sgd))):
    if i % 10 == 0:
        try:
            iteration.append(int(df_sgd_hd_adam['Iteration'][i]))
            loss_sgd_hd_adam.append(float(df_sgd_hd_adam['Loss'][i][7:-18]))
            if df_sgd_hd_adam['Alpha'][i].find('tensor') < 0:
                alpha_sgd_hd_adam.append(float(df_sgd_hd_adam['Alpha'][i]))
            else:
                alpha_sgd_hd_adam.append(float(df_sgd_hd_adam['Alpha'][i][7:-18]))

            loss_sgd_hd_nag.append(float(df_sgd_hd_nag['Loss'][i][7:-18]))
            if df_sgd_hd_nag['Alpha'][i].find('tensor') < 0:
                alpha_sgd_hd_nag.append(float(df_sgd_hd_nag['Alpha'][i]))
            else:
                alpha_sgd_hd_nag.append(float(df_sgd_hd_nag['Alpha'][i][7:-18]))

            loss_sgd_hd.append(float(df_sgd_hd['Loss'][i][7:-18]))
            if df_sgd_hd['Alpha'][i].find('tensor') < 0:
                alpha_sgd_hd.append(float(df_sgd_hd['Alpha'][i]))
            else:
                alpha_sgd_hd.append(float(df_sgd_hd['Alpha'][i][7:-18]))

            loss_sgd.append(float(df_sgd['Loss'][i][7:-18]))
            alpha_sgd.append(float(df_sgd['Alpha'][i]))

        except:
            continue
```

```python
print(len(iteration))
print(len(alpha_sgd))
```
