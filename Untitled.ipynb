```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_sgd_hd_adam = pd.read_csv('test2/logreg/+1e-03_+1e-03/sgd_hd_adam.csv')
df_sgd_hd_nag = pd.read_csv('test2/logreg/+1e-03_+1e-03/sgd_hd_nag.csv')
df_sgd_hd = pd.read_csv('test2/logreg/+1e-03_+1e-03/sgd_hd.csv')
df_sgd = pd.read_csv('test2/logreg/+1e-03_+1e-03/sgd.csv')
```

```python
iteration = []
loss_sgd_hd_adam = []
loss_sgd_hd_nag = []
loss_sgd_hd = []
loss_sgd = []
alpha_sgd_hd_adam = []
alpha_sgd_hd_nag = []
alpha_sgd_hd = []
alpha_sgd = []

for i in range(max(len(df_sgd_hd_adam),len(df_sgd_hd_nag),len(df_sgd_hd), len(df_sgd))):
    if i % 10 == 0:
        try:
            iteration.append(int(df_sgd_hd_adam['Iteration'][i]))
            loss_sgd_hd_adam.append(float(df_sgd_hd_adam['Loss'][i][7:-18]))
            if df_sgd_hd_adam['Alpha'][i].find('tensor') < 0:
                alpha_sgd_hd_adam.append(float(df_sgd_hd_adam['Alpha'][i]))
            else:
                alpha_sgd_hd_adam.append(float(df_sgd_hd_adam['Alpha'][i][7:-18]))

            loss_sgd_hd_nag.append(float(df_sgd_hd_nag['Loss'][i][7:-18]))
            if df_sgd_hd_nag['Alpha'][i].find('tensor') < 0:
                alpha_sgd_hd_nag.append(float(df_sgd_hd_nag['Alpha'][i]))
            else:
                alpha_sgd_hd_nag.append(float(df_sgd_hd_nag['Alpha'][i][7:-18]))

            loss_sgd_hd.append(float(df_sgd_hd['Loss'][i][7:-18]))
            if df_sgd_hd['Alpha'][i].find('tensor') < 0:
                alpha_sgd_hd.append(float(df_sgd_hd['Alpha'][i]))
            else:
                alpha_sgd_hd.append(float(df_sgd_hd['Alpha'][i][7:-18]))

            loss_sgd.append(float(df_sgd['Loss'][i][7:-18]))
            alpha_sgd.append(float(df_sgd['Alpha'][i]))

        except:
            continue
```

```python
print(len(iteration))
print(len(loss_sgd_hd_adam))
print(len(loss_sgd_hd_nag))
print(len(loss_sgd_hd))
print(len(loss_sgd))
print("Alpha : ")
print(len(alpha_sgd_hd_adam))
print(len(alpha_sgd_hd_nag))
print(len(alpha_sgd_hd))
print(len(alpha_sgd))
```

```python
# Averaging Loss
averaging_length = 25
cummulative_loss_sgd_hd_adam = []
cummulative_sum = 0
for i in range(len(iteration)):
    if i < averaging_length:
        cummulative_sum += loss_sgd_hd_adam[i]
        cummulative_loss_sgd_hd_adam.append(cummulative_sum/(i+1))
    else:
        cummulative_sum = cummulative_sum + loss_sgd_hd_adam[i] - loss_sgd_hd_adam[i-averaging_length]
        cummulative_loss_sgd_hd_adam.append(cummulative_sum/(averaging_length))
print(len(cummulative_loss_sgd_hd_adam))


cummulative_loss_sgd_hd_nag = []
cummulative_sum = 0
for i in range(len(iteration)):
    if i < averaging_length:
        cummulative_sum += loss_sgd_hd_nag[i]
        cummulative_loss_sgd_hd_nag.append(cummulative_sum/(i+1))
    else:
        cummulative_sum = cummulative_sum + loss_sgd_hd_nag[i] - loss_sgd_hd_nag[i-averaging_length]
        cummulative_loss_sgd_hd_nag.append(cummulative_sum/(averaging_length))
print(len(cummulative_loss_sgd_hd_nag))


cummulative_loss_sgd_hd = []
cummulative_sum = 0
for i in range(len(iteration)):
    if i < averaging_length:
        cummulative_sum += loss_sgd_hd[i]
        cummulative_loss_sgd_hd.append(cummulative_sum/(i+1))
    else:
        cummulative_sum = cummulative_sum + loss_sgd_hd[i] - loss_sgd_hd[i-averaging_length]
        cummulative_loss_sgd_hd.append(cummulative_sum/(averaging_length))
print(len(cummulative_loss_sgd_hd))

cummulative_loss_sgd = []
cummulative_sum = 0
for i in range(len(iteration)):
    if i < averaging_length:
        cummulative_sum += loss_sgd[i]
        cummulative_loss_sgd.append(cummulative_sum/(i+1))
    else:
        cummulative_sum = cummulative_sum + loss_sgd[i] - loss_sgd[i-averaging_length]
        cummulative_loss_sgd.append(cummulative_sum/(averaging_length))
print(len(cummulative_loss_sgd))
```

```python
# Averaging Alpha
averaging_length = 200
cummulative_alpha_sgd_hd_adam = []
cummulative_sum = 0
for i in range(len(iteration)):
    if i < averaging_length:
        cummulative_sum += alpha_sgd_hd_adam[i]
        cummulative_alpha_sgd_hd_adam.append(cummulative_sum/(i+1))
    else:
        cummulative_sum = cummulative_sum + alpha_sgd_hd_adam[i] - alpha_sgd_hd_adam[i-averaging_length]
        cummulative_alpha_sgd_hd_adam.append(cummulative_sum/(averaging_length))
print(len(cummulative_alpha_sgd_hd_adam))


cummulative_alpha_sgd_hd_nag = []
cummulative_sum = 0
for i in range(len(iteration)):
    if i < averaging_length:
        cummulative_sum += alpha_sgd_hd_nag[i]
        cummulative_alpha_sgd_hd_nag.append(cummulative_sum/(i+1))
    else:
        cummulative_sum = cummulative_sum + alpha_sgd_hd_nag[i] - alpha_sgd_hd_nag[i-averaging_length]
        cummulative_alpha_sgd_hd_nag.append(cummulative_sum/(averaging_length))
print(len(cummulative_alpha_sgd_hd_nag))


cummulative_alpha_sgd_hd = []
cummulative_sum = 0
for i in range(len(iteration)):
    if i < averaging_length:
        cummulative_sum += alpha_sgd_hd[i]
        cummulative_alpha_sgd_hd.append(cummulative_sum/(i+1))
    else:
        cummulative_sum = cummulative_sum + alpha_sgd_hd[i] - alpha_sgd_hd[i-averaging_length]
        cummulative_alpha_sgd_hd.append(cummulative_sum/(averaging_length))
print(len(cummulative_alpha_sgd_hd))

cummulative_alpha_sgd = []
cummulative_sum = 0
for i in range(len(iteration)):
    if i < averaging_length:
        cummulative_sum += alpha_sgd[i]
        cummulative_alpha_sgd.append(cummulative_sum/(i+1))
    else:
        cummulative_sum = cummulative_sum + alpha_sgd[i] - alpha_sgd[i-averaging_length]
        cummulative_alpha_sgd.append(cummulative_sum/(averaging_length))
print(len(cummulative_alpha_sgd))
```

```python
plt.figure(figsize=(50,25))
plt.plot(iteration, loss_sgd_hd_adam, color='red', label='sgd_hd_adam')
plt.plot(iteration, loss_sgd_hd_nag, color='blue', label='sgd_hd_nag')
plt.plot(iteration, loss_sgd_hd, color='green', label='sgd_hd')
plt.plot(iteration, loss_sgd, color='black', label='sgd')
plt.legend(loc='best')
plt.savefig('HDplots/loss.png')
```
