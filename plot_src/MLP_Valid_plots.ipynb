```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
sgd_adam = pd.read_csv('../test/mlp/+1e-03_+1e-03/sgd_HdAdam.csv')
sgd_hd = pd.read_csv('../test/mlp/+1e-03_+1e-03/sgd_Hd.csv')
sgd_sgdn = pd.read_csv('../test/mlp/+1e-03_+1e-03/sgd_Hdmomentum.csv')
sgd = pd.read_csv('../test/mlp/+1e-03_+1e-06/sgd.csv')

sgdn_hd = pd.read_csv('../test/mlp/+1e-03_+1e-03/sgdn_Hd.csv')
sgdn_sgdn = pd.read_csv('../test/mlp/+1e-03_+1e-03/sgdn_Hdmomentum.csv')
sgdn_adam = pd.read_csv('../test/mlp/+1e-03_+1e-03/sgdn_HdAdam.csv')
sgdn = pd.read_csv('../test/mlp/+1e-03_+1e-06/sgdn.csv')

adam = pd.read_csv('../test/mlp/+1e-03_+1e-06/adam.csv')
adam_hd = pd.read_csv('../test/mlp/+1e-03_+1e-07/adam_Hd.csv')
adam_adam = pd.read_csv('../test/mlp/+1e-03_+1e-07/adam_HdAdam.csv')
adam_sgdn = pd.read_csv('../test/mlp/+1e-03_+1e-07/adam_Hdmomentum.csv')
```

```python
sgd_adam.head()
```

```python
data = pd.DataFrame()
data['sgd_HdAdam'] = sgd_adam['ValidLossEpoch']
data['sgd_Hd'] = sgd_hd['ValidLossEpoch']
data['sgd_Hdmomentum'] = sgd_sgdn['ValidLossEpoch']
data['sgd'] = sgd['ValidLossEpoch']
#data['sgdn_HdAdam'] = sgdn_adam['ValidLossEpoch']
data['sgdn_Hd'] = sgdn_hd['ValidLossEpoch']
data['sgdn_Hdmomentum'] = sgdn_sgdn['ValidLossEpoch']
data['sgdn'] = sgdn['ValidLossEpoch']
data['adam_HdAdam'] = adam_adam['ValidLossEpoch']
#data['adam_Hdmomentum'] = adam_sgdn['ValidLossEpoch']
data['adam'] = adam['ValidLossEpoch']
data['adam_Hd'] = adam_hd['ValidLossEpoch']
```

```python
data.head()
```

```python
data_ = data.dropna(axis=0, how='all')
```

```python
data_.head()
```

```python
len(data_)
```

```python
a = list(data.columns)
a = [ 'sgd_Hd','sgd_Hdmomentum','sgd_HdAdam', 'sgdn_Hd', 'sgdn_Hdmomentum','adam_Hd','adam_HdAdam']
```

```python
a
```

```python
config = {}
config['sgd_HdAdam'] = {'c':'#1109ed', 'linestyle':'-.'}#03d1ff
config['sgd_Hd'] = {'c':'#1109ed', 'linestyle':'--', 'l':'op_sgd_lop_sgd (SGD-HD)'}
config['sgd_Hdmomentum'] = {'c':'#1109ed', 'linestyle':'-', 'l':'op_sgd_lop_sgdn'}#11f505
config['sgd'] = {'c':'#d5a767', 'linestyle':'-'}
#config['sgdn_HdAdam'] = {'c':'green', 'linestyle':'-'}
config['sgdn_Hd'] = {'c':'#fa0207', 'linestyle':'--', 'l':'op_sgdn_lop_sgd (SGDN-HD)'}
config['sgdn_Hdmomentum'] = {'c':'#fa0207', 'linestyle':'-', 'l':'op_sgdn_lop_sgdn'}#f705b2
config['sgdn'] = {'c':'#0339ff', 'linestyle':'-'}
config['adam_HdAdam'] = {'c':'#257002', 'linestyle':'-', 'l':'op_adam_lop_adam'}#0df705
#config['adam_Hdmomentum'] = {'c':'lime','linestyle':'--'}
config['adam'] = {'c':'#fa4c05', 'linestyle':'-'}
config['adam_Hd'] = {'c':'#257002', 'linestyle':'--', 'l':'op_adam_lop_sgd (Adam-HD)'}##11f505
data_ = data_.reset_index()
```

```python
plt.figure()
SMALL_SIZE= 18
MED_SIZE= 22
BIG_SIZE = 32
plt.rc('legend', fontsize=MED_SIZE)    # legend fontsize
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE) 
plt.rc('axes', labelsize=BIG_SIZE)    # fontsize of the x and y labels
param = {'size':'42'}
for optimizer in a:
    if optimizer not in ['adam', 'sgd', 'sgdn', 'sgd_HdAdam']:
        data_[optimizer].plot(color = config[optimizer]['c'], linestyle = config[optimizer]['linestyle'], label=config[optimizer]['l'])
    #if (optimizer in ['adam_HdAdam',
 #'adam_Hdmomentum',
 #'adam',
 #'adam_Hd' ]):
  #      data_[optimizer].plot(color = config[optimizer]['c'], l = config[optimizer]['linestyle'])

plt.legend(loc = 'best')
plt.xlabel('Epochs', **param)
plt.ylabel('Validation Loss')
plt.ylim(top = 0.40)
plt.grid(True)

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)

#plt.show()
plt.savefig('MLP_valid_Loss.pdf', bbox_inches='tight')
```
