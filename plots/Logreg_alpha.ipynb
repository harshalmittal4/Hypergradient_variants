```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
sgd_adam = pd.read_csv('test/logreg/+1e-03_+1e-03/sgd_HdAdam.csv')
sgd_hd = pd.read_csv('test/logreg/+1e-03_+1e-03/sgd_Hd.csv')
sgd_sgdn = pd.read_csv('test/logreg/+1e-03_+1e-03/sgd_Hdmomentum.csv')
sgd = pd.read_csv('test/logreg/+1e-03_+1e-06/sgd.csv')

sgdn_hd = pd.read_csv('test/logreg/+1e-03_+1e-03/sgdn_Hd.csv')
sgdn_sgdn = pd.read_csv('test/logreg/+1e-03_+1e-03/sgdn_Hdmomentum.csv')
sgdn_adam = pd.read_csv('test/logreg/+1e-03_+1e-03/sgdn_HdAdam.csv')
sgdn = pd.read_csv('test/logreg/+1e-03_+1e-06/sgdn.csv')

adam = pd.read_csv('test/logreg/+1e-03_+1e-06/adam.csv')
adam_hd = pd.read_csv('test/logreg/+1e-03_+1e-07/adam_Hd.csv')
adam_adam = pd.read_csv('test/logreg/+1e-03_+1e-07/adam_HdAdam.csv')
adam_sgdn = pd.read_csv('test/logreg/+1e-03_+1e-07/adam_Hdmomentum.csv')
```

```python
sgd_adam.head()
```

```python
data = pd.DataFrame()
data['sgd_HdAdam'] = sgd_adam['AlphaEpoch']
data['sgd_Hd'] = sgd_hd['AlphaEpoch']
data['sgd_Hdmomentum'] = sgd_sgdn['AlphaEpoch']
data['sgd'] = sgd['AlphaEpoch']
#data['sgdn_HdAdam'] = sgdn_adam['AlphaEpoch']
data['sgdn_Hd'] = sgdn_hd['AlphaEpoch']
data['sgdn_Hdmomentum'] = sgdn_sgdn['AlphaEpoch']
data['sgdn'] = sgdn['AlphaEpoch']
data['adam_HdAdam'] = adam_adam['AlphaEpoch']
#data['adam_Hdmomentum'] = adam_sgdn['AlphaEpoch']
data['adam'] = adam['AlphaEpoch']
data['adam_Hd'] = adam_hd['AlphaEpoch']
```

```python
data.head()
```

```python
data_ = data.dropna(axis=0, how='all')
```

```python
data_.head()
```

```python
data_.reset_index(inplace = True)
```

```python
len(data_)
```

```python
a = list(data.columns)
```

```python
a = [ 'sgd_Hd','sgd_Hdmomentum','sgd_HdAdam', 'sgdn_Hd', 'sgdn_Hdmomentum','adam_Hd','adam_HdAdam']
```

```python
config = {}
config['sgd_HdAdam'] = {'c':'#1109ed', 'linestyle':'-.'}#03d1ff
config['sgd_Hd'] = {'c':'#1109ed', 'linestyle':'--'}
config['sgd_Hdmomentum'] = {'c':'#1109ed', 'linestyle':'-'}#11f505
config['sgd'] = {'c':'#d5a767', 'linestyle':'-'}
#config['sgdn_HdAdam'] = {'c':'green', 'linestyle':'-'}
config['sgdn_Hd'] = {'c':'#fa0207', 'linestyle':'--'}
config['sgdn_Hdmomentum'] = {'c':'#fa0207', 'linestyle':'-'}#f705b2
config['sgdn'] = {'c':'#0339ff', 'linestyle':'-'}
config['adam_HdAdam'] = {'c':'#257002', 'linestyle':'-'}#0df705
#config['adam_Hdmomentum'] = {'c':'lime','linestyle':'--'}
config['adam'] = {'c':'#fa4c05', 'linestyle':'-'}
config['adam_Hd'] = {'c':'#257002', 'linestyle':'--'}##11f505
```

```python
plt.figure()
for optimizer in a:
    if optimizer not in ['adam', 'sgd', 'sgdn']:
        data_[optimizer].plot(color = config[optimizer]['c'], linestyle = config[optimizer]['linestyle'])
plt.legend(loc = 'best')
#plt.xlabel('Epochs')
plt.ylabel('Alpha')
plt.ylim(bottom= -0.001, top = 0.0325)

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)
#plt.axhline(y=0.2, color='r', linestyle='-')

#plt.show()
plt.savefig('./images/Logreg_Alpha.pdf', bbox_inches='tight')
```
